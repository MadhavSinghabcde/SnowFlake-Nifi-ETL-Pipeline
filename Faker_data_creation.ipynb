```python
from faker import Faker
import csv
from datetime import datetime

fake = Faker()
```

```python
def create_csv_file(num_of_records):
    current_time = datetime.now().strftime("%Y%m%d%H%M%S")
    with open(f'FakeDataset/customer_{current_time}.csv', 'w', newline='') as csvfile:
        fieldnames = ["customer_id", "first_name", "last_name", "street_address", "city", "state", "country"]
        writer = csv.DictWriter(csvfile, fieldnames = fieldnames)
        writer.writeheader()
        for i in range(num_of_records):
            writer.writerow(
                {
            'customer_id': i,
            'first_name': fake.first_name(),
            'last_name': fake.last_name(),
            'street_address': fake.street_address(),
            'city': fake.city(),
            'state': fake.state(),
            'country': fake.country()
                }
            )
```

```python
if __name__ == '__main__':
    create_csv_file(20000)
```
